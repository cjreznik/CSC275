```python
import requests
```

```python
url = 'https://itunes.apple.com/search?term=Grateful+Dead'
```

```python
response = requests.get(url)
```

```python
print(response.status_code)
```

```python
print(response.content)
```

```python
print(response.headers)
```

```python
data = response.json()
```

```python
print(data)
```

```python
import json
def pretty_print(data):
    print(json.dumps(data,indent=2))
```

```python
pretty_print(data)
```

```python
pretty_print(data['results'])
```

```python
from datetime import datetime
def year_from_date(date):
    return datetime.strptime(date,"%Y-%m-%dT%H:%M:%SZ").year
```

```python
print(data['resultCount'])
```

```python
for obj in data['results']:
    print(obj['trackName'])
```

```python
for obj in data['results']:
    print(obj['releaseDate'])
```

```python
for obj in data['results']:
    print(year_from_date(obj['releaseDate']))
```

```python
for obj in data['results']:
    year = year_from_date(obj['releaseDate'])
    print(year)
```

```python
gdYears = {}
for obj in data['results']:
    year = year_from_date(obj['releaseDate'])
    if year in gdYears.keys():
        gdYears[year] += 1
    else:
        gdYears[year]=1
print(gdYears)
```
